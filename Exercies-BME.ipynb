```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import numpy as np
from numpy.fft import *
import random

#Exercice 1
f = random.randint(0, 100) # Frequency
amplitude=random.randint(0,10)
f_s = 50  # Sampling rate, or number of measurements per second
t = np.linspace(0, 2, 2 * f_s, endpoint=False)
pure_signal = amplitude*np.sin(f*2*np.pi*t)+amplitude*np.cos(f*2*np.pi*t)

#Exercice 2
plt.plot(t,pure_signal)
plt.title('Periodic signal')
plt.ylabel('Signal amplitude')
plt.xlabel('Time(s)')
plt.show()

A = np.fft.fft(pure_signal)
plt.plot(np.real(A))
plt.show()

#Exercice 3
noise = np.random.normal(0, 3, pure_signal.shape)
signal_noise = pure_signal+ noise

plt.plot(t,signal_noise)
plt.title('Periodic signal with noise')
plt.ylabel('Signal amplitude')
plt.xlabel('Time(s)')
plt.show()

B = np.fft.fft(signal_noise)
plt.plot(np.real(B))
plt.show()

#Exercice 4
values_original = pure_signal.reshape((len(pure_signal), 1))
scaler=StandardScaler()
data_standard=scaler.fit_transform(values_original)

plt.plot(t,data_standard)
plt.title('Periodic signal standardized')
plt.ylabel('Signal amplitude')
plt.xlabel('Time(s)')
plt.show()

values_noise = signal_noise.reshape((len(signal_noise), 1))
scaler=StandardScaler()
data_standard_noise=scaler.fit_transform(values_noise)

plt.plot(t,data_standard_noise)
plt.title('Periodic signal with noise standardized')
plt.ylabel('Signal amplitude')
plt.xlabel('Time(s)')
plt.show()

#exercice 5
from sklearn.cluster import KMeans
frq1 = random.randint(0, 100)
ampl1=random.randint(0, 100)
frq2 = random.randint(0, 100)
ampl2=random.randint(0, 100)

s1= ampl1*np.sin(frq1* 2 * np.pi*t)+ampl2*np.cos(frq1*2*np.pi*t)
s2= ampl2*np.sin(frq2* 2 * np.pi*t)+ampl2*np.cos(frq2*2*np.pi*t)
clustering =KMeans(n_clusters=2)
clustering.fit([s1,s2])
print(clustering.labels_)

frq3 = random.randint(0, 100)
ampl3=random.randint(0, 100)
s3= ampl3*np.sin(frq3*2*np.pi*t)+ampl3*np.cos(frq3*2*np.pi*t)
clustering.predict([s3])



```
